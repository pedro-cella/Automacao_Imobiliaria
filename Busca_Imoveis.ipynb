## Paulo Alencar Imóveis

### Octogonal

```python
lista_imoveis = []
```

```python
# Passo 1: Acessar site
navegador = webdriver.Chrome()
navegador.get('https://www.pauloalencar.com.br/alugar/df/brasilia/area-octogonal/apartamento/valor-0_1500/')
pagina = navegador.find_element(By.CLASS_NAME, 'imoveis')
imoveis = pagina.find_elements(By.CSS_SELECTOR, '.resultado.resultado_lista.resultado_')

# Declarando lista e dicionário
lista_imoveis_octogonal = []
dados_imovel = {}

# Buscando informações
for imovel in imoveis:
    dados = imovel.find_element(By.CLASS_NAME, 'dados')
    link = imovel.find_element(By.TAG_NAME, 'a').get_attribute('href')
    dados_imovel['Link_Imovel'] = link
    dado_localizacao = dados.find_element(By.CLASS_NAME, 'localizacao')
    localizacao = dado_localizacao.find_element(By.TAG_NAME, 'span').text
    dados_imovel['Localização'] = localizacao
    dado_valor = dados.find_element(By.CLASS_NAME, 'alinha_valores')
    valor = dado_valor.find_element(By.TAG_NAME, 'h5').text
    dados_imovel['Valor'] = valor
    detalhes = dados.find_element(By.CLASS_NAME, 'detalhes').text
    dados_imovel['Detalhes'] = detalhes.replace('\n', ', ')
    lista_imoveis_octogonal.append(dados_imovel.copy())
    lista_imoveis.append(dados_imovel.copy())
    
print(lista_imoveis_octogonal)
```

```python
especial = pagina.find_element(By.CSS_SELECTOR, '.resultado.resultado_lista.resultado_last')
link = especial.find_element(By.XPATH, '//*[@id="70448334"]/a').get_attribute('href')
dados_imovel['Link_Imovel'] = link
dado_localizacao = especial.find_element(By.CLASS_NAME, 'localizacao')
localizacao = dado_localizacao.find_element(By.TAG_NAME, 'span').text
dados_imovel['Localização'] = localizacao
dado_valor = especial.find_element(By.CLASS_NAME, 'alinha_valores')
valor = dado_valor.find_element(By.TAG_NAME, 'h5').text
dados_imovel['Valor'] = valor
detalhes = especial.find_element(By.CLASS_NAME, 'detalhes').text
dados_imovel['Detalhes'] = detalhes.replace('\n', ', ')
lista_imoveis_octogonal.append(dados_imovel.copy())
lista_imoveis.append(dados_imovel.copy())
print(lista_imoveis_octogonal)
```

```python
df_paulo_octogonal = pd.DataFrame(lista_imoveis_octogonal)
df_paulo_octogonal
```

## Pires Imóveis

```python
navegador = webdriver.Chrome()
navegador.get('https://www.piresimoveis.com/imoveis/para-alugar/cruzeiro')

lista_imoveis = {}
pires_imoveis = []

imoveis = navegador.find_element(By.CSS_SELECTOR, '.row.listing-component-results')
card_imoveis = imoveis.find_elements(By.CSS_SELECTOR, '.col-sm-12.col-lg-6.box-align')
for imovel in card_imoveis:
    link = imovel.find_element(By.TAG_NAME, 'a').get_attribute('href')
    lista_imoveis['Link'] = link
    card = imovel.find_element(By.CLASS_NAME, 'card-carousel')
    valor = card.find_element(By.CLASS_NAME, 'info-left')
    valor_mes = valor.find_element(By.TAG_NAME, 'p').text
    valor_condominio_mes = 0
    try:
        valor_condominio = card.find_element(By.CSS_SELECTOR, '.info-right.text-xs-right')
        valor_condominio_mes = valor_condominio.find_element(By.TAG_NAME, 'span').text
    except:
        valor_condominio_mes = 0
    if valor_condominio_mes == 0:
        lista_imoveis['Valor'] = valor_mes
        lista_imoveis['Condomínio'] = "Sem condomínio"
    else:
        lista_imoveis['Valor'] = valor_mes
        lista_imoveis['Condomínio'] = valor_condominio_mes
    dados = imovel.find_element(By.CSS_SELECTOR, '.card-block.text-xs-center.exclusive-tag-space')
    area = dados.find_element(By.CLASS_NAME, 'card-title').text
    lista_imoveis['Área'] = area
    tipo = dados.find_element(By.CLASS_NAME, 'card-text').text
    lista_imoveis['Tipo'] = tipo
    detalhes = dados.find_element(By.CSS_SELECTOR, '.description.hidden-sm-down').text
    lista_imoveis['Detalhes'] = detalhes
    valores = dados.find_elements(By.CLASS_NAME, 'value')
    metro_quadrado = valores[-1].find_element(By.TAG_NAME, 'span').text
    lista_imoveis['m²'] = metro_quadrado
    pires_imoveis.append(lista_imoveis.copy())
print(pires_imoveis)
```

```python
df_Pires_Imoveis = pd.DataFrame(pires_imoveis)
df_Pires_Imoveis
```

```python
df_Pires_Imoveis.to_excel('pires_imoveis.xlsx', index=False)
```
