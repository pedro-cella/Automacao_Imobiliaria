```python
# Importando bilbiotecas
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import pandas as pd
```

## Paulo Alencar Imóveis

### Octogonal

```python
lista_imoveis = []
```

```python
# Passo 1: Acessar site
navegador = webdriver.Chrome()
navegador.get('https://www.pauloalencar.com.br/alugar/df/brasilia/area-octogonal/apartamento/valor-0_1500/')
pagina = navegador.find_element(By.CLASS_NAME, 'imoveis')
imoveis = pagina.find_elements(By.CSS_SELECTOR, '.resultado.resultado_lista.resultado_')

# Declarando lista e dicionário
lista_imoveis_octogonal = []
dados_imovel = {}

# Buscando informações
for imovel in imoveis:
    dados = imovel.find_element(By.CLASS_NAME, 'dados')
    link = imovel.find_element(By.TAG_NAME, 'a').get_attribute('href')
    dados_imovel['Link_Imovel'] = link
    dado_localizacao = dados.find_element(By.CLASS_NAME, 'localizacao')
    localizacao = dado_localizacao.find_element(By.TAG_NAME, 'span').text
    dados_imovel['Localização'] = localizacao
    dado_valor = dados.find_element(By.CLASS_NAME, 'alinha_valores')
    valor = dado_valor.find_element(By.TAG_NAME, 'h5').text
    dados_imovel['Valor'] = valor
    detalhes = dados.find_element(By.CLASS_NAME, 'detalhes').text
    dados_imovel['Detalhes'] = detalhes.replace('\n', ', ')
    lista_imoveis_octogonal.append(dados_imovel.copy())
    lista_imoveis.append(dados_imovel.copy())
    
print(lista_imoveis_octogonal)
```

```python
especial = pagina.find_element(By.CSS_SELECTOR, '.resultado.resultado_lista.resultado_last')
link = especial.find_element(By.XPATH, '//*[@id="70448334"]/a').get_attribute('href')
dados_imovel['Link_Imovel'] = link
dado_localizacao = especial.find_element(By.CLASS_NAME, 'localizacao')
localizacao = dado_localizacao.find_element(By.TAG_NAME, 'span').text
dados_imovel['Localização'] = localizacao
dado_valor = especial.find_element(By.CLASS_NAME, 'alinha_valores')
valor = dado_valor.find_element(By.TAG_NAME, 'h5').text
dados_imovel['Valor'] = valor
detalhes = especial.find_element(By.CLASS_NAME, 'detalhes').text
dados_imovel['Detalhes'] = detalhes.replace('\n', ', ')
lista_imoveis_octogonal.append(dados_imovel.copy())
lista_imoveis.append(dados_imovel.copy())
print(lista_imoveis_octogonal)
```

```python
df_paulo_octogonal = pd.DataFrame(lista_imoveis_octogonal)
df_paulo_octogonal
```

### Sudoeste

```python
# Passo 1: Acessar site
navegador = webdriver.Chrome()
navegador.get('https://www.pauloalencar.com.br/alugar/df/brasilia/setor-sudoeste/apartamento/ordem-valor/resultado-crescente/quantidade-12/')
pagina = navegador.find_element(By.CLASS_NAME, 'imoveis')
imoveis = pagina.find_elements(By.CSS_SELECTOR, '.resultado.resultado_lista.resultado_')
imoveis_last = pagina.find_elements(By.CSS_SELECTOR, '.resultado.resultado_lista.resultado_last')

# Declarando lista e dicionário
lista_imoveis_sudoeste = []
dados_imovel = {}

for imovel in imoveis:
    dados = imovel.find_element(By.CLASS_NAME, 'dados')
    link = imovel.find_element(By.TAG_NAME, 'a').get_attribute('href')
    dados_imovel['Link_Imovel'] = link
    dado_localizacao = dados.find_element(By.CLASS_NAME, 'localizacao')
    localizacao = dado_localizacao.find_element(By.TAG_NAME, 'span').text
    dados_imovel['Localização'] = localizacao
    dado_valor = dados.find_element(By.CLASS_NAME, 'alinha_valores')
    valor = dado_valor.find_element(By.TAG_NAME, 'h5').text
    dados_imovel['Valor'] = valor
    detalhes = dados.find_element(By.CLASS_NAME, 'detalhes').text
    dados_imovel['Detalhes'] = detalhes.replace('\n', ', ')
    lista_imoveis_sudoeste.append(dados_imovel.copy())
    lista_imoveis.append(dados_imovel.copy())
    
for imovel in imoveis_last:
    dados = imovel.find_element(By.CLASS_NAME, 'dados')
    link = imovel.find_element(By.TAG_NAME, 'a').get_attribute('href')
    dados_imovel['Link_Imovel'] = link
    dado_localizacao = dados.find_element(By.CLASS_NAME, 'localizacao')
    localizacao = dado_localizacao.find_element(By.TAG_NAME, 'span').text
    dados_imovel['Localização'] = localizacao
    dado_valor = dados.find_element(By.CLASS_NAME, 'alinha_valores')
    valor = dado_valor.find_element(By.TAG_NAME, 'h5').text
    dados_imovel['Valor'] = valor
    detalhes = dados.find_element(By.CLASS_NAME, 'detalhes').text
    dados_imovel['Detalhes'] = detalhes.replace('\n', ', ')
    lista_imoveis_sudoeste.append(dados_imovel.copy())
    lista_imoveis.append(dados_imovel.copy())
    
print(lista_imoveis_sudoeste)
```

```python
df_paulo_sudoeste = pd.DataFrame(lista_imoveis_sudoeste)
df_paulo_sudoeste
```
